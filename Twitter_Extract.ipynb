```python
import pandas as pd
import datetime
from tqdm.notebook import tqdm # To display status bar
import snscrape.modules.twitter as sntwitter # To scrape tweets
```

```python
# Query
cp_query = "canada post"             # keyword
# query = "#canadapost"      # hashtag
# query = "@canadapostcorp"     # mention

# Language
language = "en"

cp_scraper = sntwitter.TwitterSearchScraper(cp_query + " lang:" + language)
```

```python
cp_tweets = []
cp_tweet_count = 15000

for i, tweet in tqdm(enumerate(cp_scraper.get_items()), total = cp_tweet_count + 1):
    data = [
        tweet.date, 
        tweet.id, 
        tweet.rawContent, 
        tweet.user.username, 
        tweet.likeCount, 
        tweet.retweetCount ]
    
    cp_tweets.append(data)
    if i >= cp_tweet_count:
        break
```

```python
cp_tweet_df = pd.DataFrame(cp_tweets,  columns = ['date', 'id', 'content', 'username', 'like_count', 'retweet_count'])
```

```python
cp_tweet_df.to_csv('twitter_data.csv', index=False)
```
